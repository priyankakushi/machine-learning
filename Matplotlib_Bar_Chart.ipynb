```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
# Dataset of "indian artificial intelligence production (IAIP) class"
#classes = ["python", "R", "artificial intelligence", "machine learning", "data science"]
classes = ["python", "R", "AI", "ML", "DS"]
class1_students = [30, 10, 20, 25, 10]
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 15]
```

```python
plt.bar(classes, class1_students, width = 0.2, align = "edge", color = "g",
       edgecolor="m", linewidth=5, alpha=0.9, linestyle = "--")
plt.xlabel("classes")
plt.ylabel("No of students")
plt.title("Student Classes bar chart")
plt.legend()
plt.show()
```

```python
plt.barh(classes, class1_students)
```

```python
#visible Method
plt.bar(classes, class1_students, width = 0.2, align = "edge", color = "g",
       edgecolor="m", linewidth=5, alpha=0.9, linestyle = "--", visible=False)
plt.xlabel("classes")
plt.ylabel("No of students")
plt.title("Student Classes bar chart")
plt.legend()
plt.show()
```

```python

plt.bar(classes, class1_students, width = 0.60, align = "edge", color = "g",
       edgecolor="m", linewidth=5, alpha=0.9, linestyle = "--")
style.use("ggplot")
plt.xlabel("classes")
plt.ylabel("No of students")
plt.title("Student Classes bar chart")

plt.legend()
plt.show()
```

```python
plt.bar(classes, class1_students, width = 0.60, align = "edge", color = "g",
       edgecolor="m", linewidth=5, alpha=0.9, linestyle = "--")

plt.xlabel("classes", fontsize=15)
plt.ylabel("No of students", fontsize=15)
plt.title("Student Classes bar chart", fontsize=18)
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.bar(classes, class1_students, width = 0.6, align = "edge", color = "g",
       edgecolor="m", linewidth=5, alpha=0.9, linestyle = "--")

plt.xlabel("classes", fontsize=15)
plt.ylabel("No of students", fontsize=15)
plt.title("Student Classes bar chart", fontsize=18)
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.bar(classes, class1_students, width = 0.2, color = "b",label="class 1 student")

plt.bar(classes, class2_students, width = 0.2,  color = "g", label="class 2 student")

plt.xlabel("classes", fontsize=15)
plt.ylabel("No of students", fontsize=15)
plt.title("Student Classes bar chart", fontsize=18)
plt.show()
```

```python
plt.figure(figsize=(10,7))
classes_index = np.arange(len(classes))
width = 0.2
plt.bar(classes_index, class1_students, width, color = "b",label="class 1 student")

plt.bar(classes_index+width, class2_students, width,  color = "g", label="class 2 student")
plt.xticks(classes_index + width, classes)
plt.xlabel("classes", fontsize=15)
plt.ylabel("No of students", fontsize=15)
plt.title("Student Classes bar chart", fontsize=18)
plt.show()
```

```python
plt.figure(figsize=(10,7))
classes_index = np.arange(len(classes))
width = 0.2
plt.bar(classes_index, class1_students, width, color = "b",label="class 1 student")

plt.bar(classes_index+width, class2_students, width,  color = "g", label="class 2 student")

plt.bar(classes_index+width+width, class3_students, width,  color = "y", label="class 3 student")

plt.xticks(classes_index + width, classes, rotation=30)
plt.xlabel("classes", fontsize=15)
plt.ylabel("No of students", fontsize=15)
plt.title("Student Classes bar chart", fontsize=18)
plt.legend()
plt.show()
```
