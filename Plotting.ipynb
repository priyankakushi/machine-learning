# SubPlottingÂ¶

```python
data1 = np.arange(10)
data2 = np.array([0, 2, 2, 6, 4, 10, 6, 14, 8, 18])
data3 = np.array(np.random.randint(0, 18, 20))
data4 = np.array(np.random.randint(0, 100, 50))
print(data1)
print(data2)
print(data3)
print(data4)
```

```python
#plotting data 1
plt.plot(data1, "k-",label = "default")
plt.show()
```

```python
##ploting data2
plt.plot(data2,"r--*", label = "steps")
plt.show()
```

```python
##plotting data3
# len(data3) check lenth of data3
x = np.arange(1,21)
plt.scatter(x,data3,label = "Scater", color = "g", s=60, marker = "*", alpha = 0.9)
plt.show()
```

```python
plt.hist(data4, bins = 6, alpha = 0.9, color = "g", edgecolor = "k", label = "Hist")
plt.show()
```

```python
fig = plt.figure(figsize = (8, 8))
fig.subplots_adjust(hspace=0.2, wspace = 0.3)
ax1 = fig.add_subplot(2,2,1)##figure should have 2x2 or utp 4 fig, this is figure
ax1.plot(data1, "k-", label = "Line")
plt.legend()
plt.xlabel("stages")
plt.ylabel("values")
###################################################
ax2 = fig.add_subplot(2,2,2)
ax2.plot(data2, "r--*", label = "step")
plt.legend()
plt.xlabel("stages")
plt.ylabel("values")
##############################################
ax3=fig.add_subplot(2,2,3)
ax3.scatter(np.arange(20), data3, label="Scater", color = "g", s=25, marker="*", alpha=0.9)
plt.legend(loc = "best")
plt.xlabel("stages")
plt.ylabel("values")
######## Histogram plot ##########
ax4 = fig.add_subplot(2,2,4)
ax4.hist(data4,bins=10,alpha=0.9,histtype="stepfilled",color="g",edgecolor="r", label="hist")
plt.legend(loc="best")
plt.xlabel("stages")
plt.ylabel("values")
plt.show()
plt.close()
```
