# Combining and Merging Datasets

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({"key":["b", "b", "a", "c", "a", "a", "b"], "data1":range(7)})
df2 = pd.DataFrame({"key": ["a", "b", "d"], "data2": range(3)})
print(df1)
print(df2)
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1,df2, on="key")
```

```python
df1 = pd.DataFrame({"Student":["John", "John", "Adam", "Smith", "Adam", "Adam", "John"], "Absentism":[10,4,5,7,3,2,6]})
df2 = pd.DataFrame({"Student":["Adam", "Bruce", "John"], "Absentism":[0, 1, 2]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2, on = "Student")
```

```python
pd.merge(df1, df2, on = "Student", how = "outer")
```

```python
dff1 = pd.DataFrame({"Student":["John", "John", "Adam", "Smith", "Adam", "Adam", "John"], "Absentism":[10,4,5,7,3,2,6]})
dff2 = pd.DataFrame({"Student":["John","John", "Adam", "Smith", "Adam", "Adam", "John"], "Grade":[10,4,5,7,3,2,6]})
```

```python
dff1
```

```python
dff2
```

```python
pd.merge(dff1, dff2, on = "Student", how = "outer")
#for union of both the
```

```python
pd.merge(dff1, dff2, on = "Student", how = "inner")
```

```python
D1 = pd.DataFrame({"key": ["b", "b", "a", "c", "a", "b"], "data1": range(6)})
D2 = pd.DataFrame({"key": ["a", "b", "a", "b", "d"], "data2": range(5)})
print("Frist DataFrame: \n", D1, "\n")
print("Second DataFrame: \n", D2, "\n")
```

```python
print(pd.merge(D1, D2, on = "key", how = "left"))
print(pd.merge(D1, D2, on = "key", how = "right"))
```

# To merge with multiple keys, pass a list of column names:

```python
left = pd.DataFrame({"key1": ["foo", "foo", "bar"],
                    "key2": ["one", "two", "one"],
                    "lval": [1,2,3]})
right = pd.DataFrame({"key1": ["foo", "foo", "bar", "bar"],
                     "key2": ["one", "one", "one", "two"],
                     "rval": [4,5,6,7]})
result = pd.merge(left, right, on = ["key1", "key2"], how = "outer")
print(left)
print(right)
print(result)
```

```python
#in this case, you can pass : left_index = True OR right_index = True or both.
```

```python
df_1 = pd.DataFrame({"key": ["a", "b", "a", "a", "b", "c"], "value":range(6)})
df_r = pd.DataFrame({"group_val": [3.5,7]}, index = ["a", "b"])
```

```python
df_1
```

```python
df_r
```

```python
pd.merge(df_1, df_r, left_on = "key", right_index=True)
```

```python
pd.merge(df_1, df_r, left_on = "key", right_index=True, how="outer")
```

# Concatenating along an axis

```python
#note: conctenating arguments axis, join, join_axes, keys,levels, names, ignore_index
```

```python
import pandas as pd
import numpy as np
```

```python
s1 = pd.Series([0,1], index=["b", "a"])
s2 = pd.Series([2,3,4], index = ["c", "d", "e"])
s3 = pd.Series([5,6], index=["f", "g"])
print('s1\n', s1, "\n")
print('s2\n', s2, "\n")
print("s3\n",s3)
```

```python
pd.concat([s1, s2, s3], axis=1, sort=False)
```

```python
pd.concat([s1,s2,s3], axis=1, sort=True)
```

```python
pd.concat([s1,s2,s3], axis=0, sort=True)
```

```python
pd.concat([s1,s2,s3], axis=0, sort=False)
```

```python
s4 = pd.Series([0,1,5,6], index = ["b", "a", "f", "g"])
s4
```

```python
pd.concat([s1, s4], axis=1, sort=True)
```

```python
pd.concat([s1, s4], axis=1, join="inner", sort=True)
```

```python
pd.concat([s1,s4], axis=1, join = "outer", sort=True)
```

```python
pd.concat([s1,s4],axis=1, join_axes=[["a", "c", "b", "e"]])
```

```python
result = pd.concat([s1, s2, s3], keys = ["s1", "s2", "s3"])
result
```

```python
result.unstack()
```

```python
result.unstack().T
```

```python
result.unstack().stack()
```

# DataFrame: same logic is extended in DataFrame

```python
df1 = pd.DataFrame(np.arange(6).reshape(3,2),
                  index=["a", "b", "c"], columns=["one", "two"])
df2 = pd.DataFrame(np.arange(4).reshape(2,2),
                  index=["a", "b"], columns=["three", "four"])
print(df1)
print(df2)
```

```python
pd.concat([df1,df2],keys=["level1", "level2"], sort=False)
```

```python
pd.concat([df1,df2],axis=1, keys=["level1", "level2"], sort=True)
```

```python
pd.concat([df1,df2], axis = 1, keys=["df1", "df2"], sort=True)
```

```python
pd.concat({"df1":df1, "df2":df2},axis=1,sort=True)
```

```python
pd.concat([df1, df2], keys=["df1", "df2"], names=["upper", "lower"], axis=1,sort=False)
```

```python
dff1 = pd.DataFrame(np.random.randn(3,4), columns=["a", "b", "c", "d"])
dff2 = pd.DataFrame(np.random.randn(2,3), columns=["b", "d", "c"])
print(dff1, "\n")
print(dff2)
```

```python
pd.concat([dff1,dff2], sort=True)
```

```python
pd.concat([dff1,dff2], ignore_index = True, sort=True)
```
